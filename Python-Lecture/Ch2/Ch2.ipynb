#Python Data Types


*   Numeric type: int & float  --> immutable (cannot change or update)
*   Text type: str (string)  -> immutable (cannot change or update)
*   Sequence Type: list (mutable), tuple, range
*   Mapping type: dict (mutable)



```python
x = 10
#Object has 1. value 2. Type 3. Idenity

print(" x", x,  "type", type(x), "id", id(x))
```

```python
y = 3.14
#object of type/class float
print ("y", y, "type", type(y), "id", id(y))
```

```python
#int/float is immutable
x = 10
x = 5
print (x)
```

```python
#String  --> immutable

first_name = "Kevin"
first_name = "Alice"

```
